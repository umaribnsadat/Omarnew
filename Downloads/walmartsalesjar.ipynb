```python
import pandas as pd

file_path = r'C:\Users\umarh\Downloads\Walmart Sales.xlsx'
data = pd.read_excel(file_path)

data['Date'] = pd.to_datetime(data['Date'])

if 'Unit price' in data.columns and 'Quantity' in data.columns:
    data['Revenue'] = data['Unit price'] * data['Quantity']

# Task B: Average Price of an Item Sold at Each Branch of the City
average_price_per_branch = data.groupby(['City', 'Branch'])['Unit price'].mean().reset_index()
print("Average Price of an Item Sold at Each Branch of the City:")
print(average_price_per_branch)

# Task C: Month over Month Analysis
data['Month-Year'] = data['Date'].dt.to_period('M')

# Grouping data by 'Month-Year'
monthly_performance = data.groupby(['Month-Year', 'Product line', 'Gender', 'Payment'])['Revenue'].sum().reset_index()
print("\nMonth over Month Revenue Performance:")
print(monthly_performance)

april_2019_performance = monthly_performance[monthly_performance['Month-Year'] == '2019-04']
print("\nFocus Areas for Better Sales in April 2019:")
print(april_2019_performance)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder


file_path = r'C:\Users\umarh\Downloads\Walmart Sales.xlsx'


data = pd.read_excel(file_path)


if 'Product line' in data.columns:
    one_hot = pd.get_dummies(data['Product line'])
    data = pd.concat([data, one_hot], axis=1)
    data.drop('Product line', axis=1, inplace=True)
numerical_features = data.select_dtypes(include=['float64', 'int64'])
scaler = StandardScaler()
scaled_features = scaler.fit_transform(numerical_features)
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(scaled_features)
data['Cluster'] = clusters
centers = pd.DataFrame(scaler.inverse_transform(kmeans.cluster_centers_), columns=numerical_features.columns)
print("Cluster Centers:")
print(centers)
print("\nData with Cluster Labels:")
print(data.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
file_path = r'C:\Users\umarh\Downloads\Walmart Sales.xlsx'
data = pd.read_excel(file_path)
if 'Unit price' in data.columns and 'Quantity' in data.columns:
    data['Revenue'] = data['Unit price'] * data['Quantity']

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.histplot(data['Revenue'], bins=30, kde=True)
plt.title('Distribution of Revenue')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.show()

# Bar chart 
plt.figure(figsize=(12, 7))
sns.countplot(y='Product line', data=data, order = data['Product line'].value_counts().index)
plt.title('Sales Count per Product Line')
plt.xlabel('Counts')
plt.ylabel('Product Line')
plt.show()

# Box plot 
plt.figure(figsize=(10, 6))
sns.boxplot(x='Branch', y='Revenue', data=data)
plt.title('Revenue Distribution per Branch')
plt.xlabel('Branch')
plt.ylabel('Revenue')
plt.show()

# Scatter plot of Unit price vs Quantity
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Unit price', y='Quantity', hue='Branch', data=data)
plt.title('Unit Price vs Quantity per Branch')
plt.xlabel('Unit Price')
plt.ylabel('Quantity')
plt.show()

# Line plot of Revenue over Time (assuming Date is parsed correctly)
data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace=True)
data['Cumulative Revenue'] = data['Revenue'].cumsum()
plt.figure(figsize=(14, 7))
sns.lineplot(x='Date', y='Cumulative Revenue', data=data)
plt.title('Cumulative Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Revenue')
plt.show()

# Interactive Plot using Plotly
fig = px.pie(data, names='Branch', values='Revenue', title='Revenue Share by Branch')
fig.show()

# Advanced: Heatmap of correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
